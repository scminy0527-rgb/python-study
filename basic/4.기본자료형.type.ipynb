기본 자료형(Data type)
- 숫자 자료형(int)
    - 양의 정수, 음의정수, 0   0, 1, 100, ...
- 숫자 자료형(float)
    - 소수점이 있는 수      0.0, 3.14, ...
- 논리 자료형(bool)
    - 참 또는 거짓      True, False, 값이 있을 때
- 문자열 자료형(str)
    - 문자열         '', "", ...

```python
import sys
```

```python
# sys.getsizeof(): 메모리를 차지하는 바이트 확인
```

```python
# 단위가 바이트
print(sys.getsizeof(100))
print(sys.getsizeof(0.5))
print(sys.getsizeof("홍길동"))
print(sys.getsizeof(True))
```

```python
data1 = 0.5
data2 = 10
data3 = "홍길동"
data4 = False
```

```python
# 자동형변환
# 정수 + 실수 = 실수
print(data1 + data2)
```

```python
# 단 자바 처럼 정수 + 문자열 은 안됨
```

```python
# 문자형 + 논리형
print(data3 + data4)
```

```python
# 논리형 + 정수
# False: 0 / True: 1 로 보기 가능
print(True + 0.5)
```

```python
# 강제 형변환
# 홍길동 + 20
# 타입(바꿀 값): 강제형변환

int(True) 
print(type(int(True)))
```

# 인덱싱(indexing)
- 시퀀스 자료구조에서 인덱스로 값을 접근하는 문법

```python
# 문자열 도 리스트
# 인덱싱

data = 'a''b''c''d'
print(data[1])
```

## 슬라이싱 (slicing)
- 시퀀스 자료 구조에서 인덱스 범위에서 인덱스 범위까지 접근하는 문법

```python
data = "이규혁노민균문윤성노규호"
```

```python
data[-3:]
```

```python
data[:-3]
```

```python
# 스텝 사용
numbers = "123456789"
numbers[1::2]
```

```python
numbers[::-1]
```

```python
# 홀수 가져 와서 반대로 뒤집기
numbers[::-2]
```

```python
# 인덱싱과 형변환을 이용해서 -10 만들기
numbers
total = 0
for x in numbers[:4]:
    total += int(x)

print("-" + str(total))
```

```python
numbers[::-11]
```
