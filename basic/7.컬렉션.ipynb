컬렉션(list, tuple, set, dict) 메서드
1. 리스트
   리스트명.메소드명()
   .append(값)
      - 값을 리스트의 맨 뒤에 추가 해준다
   .insert(인덱스, 값)   
      - 값을 해당 idx에 추가해준다
   .extend([])
      - 리스트와 다른 리스트를 연결해준다.
   .remove(값)
      - 값을 삭제한다. (값이 없을 때는 오류가 발생)
   .pop(인덱스)
      - 해당 인덱스의 요소를 삭제한다
   .sort()
      -요소를 오름차순 정렬해준다(리스트 자체를 내부적으로 정렬)
   .sort( , reverse = True) 
      - 요소를 내림차순 정렬해준다
   sorted()
      - 리스트의 정렬된 복사본을 반환

   len() : 리스트의 길이(요소의 갯수) 알려준다


2. 튜플의 메서드
   .count(값)
      - 해당 값이 몇 개인지 값을 반환한다
   .index(값)
      - 해당 값의 인덱스 번호를 알려준다

3. 세트의 메서드
   .add(값)
      - 값을 추가한다. 중복된 값은 추가되지 않는다
   .remove(값)
      - 값을 삭제한다. 없는 요소를 넣으면 오류가 발생한다
   .discard(값)
      - 값을 삭제한다. 없는 요소를 넣어도 오류가 발생하지 않는다
   .pop()
      - 요소 한개를 삭제한다.


4. 딕셔너리의 메서드
   - 키(key)와 값(Value)가 항상 한쌍으로 저장된다
   - key값의 중복이 불가능하다

   .keys() : key값들만 모여있는 새로운 컬렉션
   .values() : value값들만 모여있는 새로운 컬렉션
   .items() : (key, value) 튜플들이 담겨있는 새로운 컬렉션
