# 4. 누적복합연산자

```python
money = 200000000
```

```python
money /= 2000
print(money)
```

```python
money *= 0.7
print(money)
```

```python
# 2진수
print(bin(10))
```

```python
# 16진수
print(hex(16))
```

```python
print(bin(10))
print(bin(11))
```

```python
# 논리합 (2진수 비교 해서 하나라도 1 이면 1 해서 하기)
print(10 | 11)
```

```python
# 논리곱, (둘 다 1 이면 1, 아니면 0)
print(11 & 10)
```

```python
# xor (모두 같으면 0 아니면 1)
print((11 ^ 10))
```

```python
# not ~
print(~10)
```
