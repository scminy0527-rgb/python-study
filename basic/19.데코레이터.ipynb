```python
from datetime import datetime

datetime.now()

def add_number(num1, num2):
    print(datetime.now())
    return num1 + num2
```

```python
add_number(10, 20)
```

```python
def concat(*args):
    return "".join(map(lambda data : str(data), args))
```

```python
concat(10, 20, 30)
```

```python
def log_time(original_function):
    def logging(*args):
        print(datetime.now())
        return original_function(*args)
    return logging
```

```python
@log_time
def add_number(num1, num2):
    return num1 + num2
```

```python
add_number(10, 30)
```

```python
def divide(num1, num2):
    try:
        return int(num1) / int(num2)
    except ZeroDivisionError:
        print("0으로 나눌 수 없습니다.")
    except ValueError:
        print("정수만 입력 가능합니다.")
    except Exception:
        print("알 수 없는 오류")
```

```python
divide(10, "r")
```

```python
def substract(num1, num2):
    try:
        return num1 - num2
    except TypeError:
        print("정수만 계산할 수 있습니다.")
```

```python
substract("d", 7)
```

```python
def safe_calc(original_function):
    def exception_wrapper(*args):
        try:
            return original_function(*args)
        except ZeroDivisionError:
            print("0으로 나눌 수 없습니다.")
        except ValueError:
            print("정수만 입력 가능합니다.")
        except TypeError:
            print("정수만 계산할 수 있습니다.")
        except Exception:
            print("알 수 없는 오류")
    return exception_wrapper
```

```python
@safe_calc
def divide(num1, num2):
    return int(num1) / int(num2)
```

```python
divide(0, "r")
```

```python
@safe_calc
def substract(num1, num2):
    return num1 - num2
```

```python
substract("d", 5)
```

```python
substract(10, 5)
```

```python
# 이름과 정수를 몇 개 입력할 지 모를 때
# "홍길동님이 전달한 총 합: 000"으로 출력해주는 메서드 구현
```

```python
def check_num(original_function):
    def exception_wrapper(*args):
        try:
            return original_function(*args)
        except ZeroDivisionError:
            print("0으로 나눌 수 없습니다.")
        except ValueError:
            print("정수만 입력 가능합니다.")
        except TypeError:
            print("정수만 계산할 수 있습니다.")
        except Exception:
            print("알 수 없는 오류")
    return exception_wrapper
```

```python
@check_num
def get_name_add_num():
    name = input("이름을 입력하세요")
    total = 0
    idx = 0
    while True:
        num = input(f"숫자를 입력하세요 {idx + 1}번째 입력 \n 종료 하시려면 q 를 입력")
        if num == "q" : break
        total += int(num)
        idx += 1
        print(f"누적합: {total}")

    print(f"{name}님이 입력한 숫자의 총 합: {total}")
        
```

```python
get_name_add_num()
```
