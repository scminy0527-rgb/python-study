```python
# 이름(name), 나이(age), 운동종목(sports)
# 플레이어들을 소개하는 메서드

# 운동선수 3명의 객체화 하고
# 3명을 각각 소개하는 메서드를 사용해서 정보들을 출력하세요.
# ex) OOO선수의 나이 OO살 운동종목 OOOO
class Player:
    def __init__(self, name, age, sports):
        self.name = name
        self.age = age
        self.sports = sports

    def introduce(self):
        print(f"{self.name}선수의 나이 {self.age}살 운동종목 {self.sports}")
        

player1 = Player("류현진", 30, "야구선수")
player2 = Player("손흥민", 30, "축구선수")
player3 = Player("이규혁", 30, "코딩선수")

player1.introduce()
player2.introduce()
player3.introduce()
```

```python
# 완전 객체 지향형 언어
# Market
# 판매
class Market:
    def __init__(self, market_name, product_name, product_price):
        self.market_name = market_name
        self.product_name = product_name
        self.product_price = product_price

    def sell(self, customer):
        customer.money -= self.product_price * (100 - customer.discount) / 100

# Customer
# 할인율
class Customer:
    def __init__(self, name, money, discount):
        self.name = name
        self.money = money
        self.discount = discount
```

```python
cu = Market("CU", "도시락", 10000)
customer = Customer("이규혁", 50000, 10)

cu.sell(customer)
print(customer.money)
```

```python
# Atm, User 객체
# deposit

# Atm.deposit(User, 10000)
# 유저의 돈은 10000 차감,
# 은행에 돈은 10000 입금

# 은행이 가진돈과 유저의 돈 출력
```

```python
class Atm:
    def __init__(self, name):
        self.name = name
        self.users = {}

    def deposit(self, user, money):

        if user.money < money: 
            return "잔액 부족"
        
        if user.name in self.users:
            # 기존의 돈
            self.users.update({user.name: self.users[user.name] + money})
            
        else:
            self.users.update({user.name: money})

        user.money -= money

    def show(self):
        print(self.users)
```

```python
class User:
    def __init__(self, name, money):
        self.name = name
        self.money = money

    def show(self):
        format_money = format(self.money, "_")
        print(f"{self.name}님의 잔 액: {format_money}원")
```

```python
yeoksam_atm = Atm("yeoksam")

ggyu = User("이규학", 50000)

yeoksam_atm.deposit(ggyu, 10000)
yeoksam_atm.deposit(ggyu, 10000)
yeoksam_atm.deposit(ggyu, 500000)
yeoksam_atm.show()
ggyu.show()
```

```python
# 클래스 매서드
class Korean:
    # 클래스 변수
    country = "한국"

    # 클래스 매서드
    # 인스턴스 변수는 사용 안하지만 클래스 변수는 사용 하는 건 클래스 매서드
    @classmethod # 데코레이터
    def trip(cls, country):
        if cls.country == country:
            print("국내여행")
        else :
            print("해외여행")

    # 정적 매서드 (컴파일러가 가장 먼저 메모리에 올리는 매서드, 따라서 클래스를 힙 메모리에 안올리더라도 바로 사용 가능)
    @staticmethod
    def slogan():
        print("Imagine your korea")
```

```python
hong = Korean()
hong.trip("한국")
hong.trip("일본")
```

```python
hong.slogan()
Korean.slogan()
```

# 상속 (Inheritance)

```python
class Person:
    def __init__(self, name):
        self.name = name

# 부모 상속
class Student(Person):
    def __init__(self, name, age):
        super().__init__(name)
        self.age = age
    
    
```

```python
hong = Person("홍길동")
print(id(hong)) # 이제 메모리 할당 주소값

jang = Student("장보고", 20)

# 부모 -> 타입도 같이 전달
# 모든 자식은 부모 타입
# 부모는 자식 타입이 아니다.
print(isinstance(hong, Person))
print(isinstance(hong, Student))
print(isinstance(jang, Person))
print(isinstance(jang, Student))
```

```python
class Animal:
    def __init__(self, name, feed, species):
        self.name = name
        self.feed = feed
        self.species = species

    def feeding(self):
        print(f"{self.name}은 먹이를 먹습니다.")
        

class Dog(Animal):
    def __init__(self, name, feed, species):
        super().__init__(name, feed, species)
    
    def act(self):
        print(f"{self.name}은 산책을 합니다.")

class Cat(Animal):
    def __init__(self, name, feed, species):
        super().__init__(name, feed, species)

    def sleep(self):
        print(f"{self.name}은 잠을 자주 잡니다.")
```

```python
dog = Dog("누렁이", "잡식", "강아지")
dog.act()

cat = Cat("야옹이", "사료", "고양이")
cat.sleep()
```

```python
# 사람(부모), 학생(자식)
# 사람 - 이름, 나이, 키, 몸무게
# 학생 - 학생, 핸드폰
class Person:
    def __init__(self, name, age, height, weight):
        self.name = name
        self.age = age
        self.height = height
        self.weight = weight

    def greet(self):
        print(f"사람 {self.name}은 인사를 합니다.")

# lee = Person("이규학", 20, 170, 60)
# lee.greet()


class Student(Person):
    def __init__(self, name, age, height, weight, student_num, phone_num):
        super().__init__(name, age, height, weight)
        self.student_num = student_num
        self.phone_num = phone_num

    def study(self):
        print(f"학생 {self.name}은 공부를 합니다.")

# 사람은 인사하는 메서드
# 학생은 공부하는 메서드
```

```python
lee = Person("이순신", 20, 170, 60)
lee.greet()

lgh = Student("이규학", 25, 170, 60, "b611108", "010-1234-1234")
lgh.study()
```

```python
class Car:
    def __init__(self, brand, price):
        self.brand = brand
        self.price = price

    def start_engine(self):
        print(f"{self.brand}는 열쇠로 시동을 겁니다.")

    def stop_engine(self):
        print(f"{self.brand}는 열쇠로 시동을 끕니다.")

class SuperCar(Car):
    def __init__(self, brand, price):
        super().__init__(brand, price)

    def start_engine(self):
        print(f"{self.brand}는 음성으로 시동을 겁니다.")

    def stop_engine(self):
        print(f"{self.brand}는 음성으로 시동을 끕니다.")
```

```python
matiz = Car("마티즈", 5000000)
benz = SuperCar("벤츠 cls", 30000000)

matiz.start_engine()
matiz.stop_engine()

benz.start_engine()
benz.stop_engine()
```

```python
# 휴대폰 필드
# 브랜드(brand), 가격(price)
# 2G 전화하기
# 3G 전화하기, 인터넷
# 4G 전화하기, 인터넷, 영상통화

class Phone2G:
    def __init__(self, brand, price):
        self.brand = brand
        self.price = price

    def call(self):
        print(f"{self.brand}은 일반 전화")

class Phone3G(Phone2G):
    def __init__(self, brand, price):
        super().__init__(brand, price)

    def call(self):
        print(f"{self.brand}은 영상 통화도 가능")

    def browse(self):
        print("인터넷 느리게 사용 가능")

class Phone4G(Phone3G):
    def __init__(self, brand, price):
        super().__init__(brand, price)

    def browse(self):
        print("인터넷 사용 가능(속도 개선)")

    def stream_vedio(self):
        print("영상 통화 가능")
```

```python
magic_hole = Phone2G("애니콜 매직홀", 400000)
magic_hole.calling()

iphone_3gs = Phone3G("아이폰 3gs", 900000)
iphone_3gs.calling()
iphone_3gs.internet()

galaxy_s2 = Phone4G("갤럭시 s2", 950000)
galaxy_s2.calling()
galaxy_s2.internet()
galaxy_s2.video_call()
```
