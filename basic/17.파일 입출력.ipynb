## with 문
- 사용한 리소스를 자동으로 닫아주는 문법(관리해주는 문법)

```python
with open("./food_list.txt", "w", encoding="utf-8") as f:
    foods = ["돈까스", "제육볶음", "김치찌개", "순대국밥", "오징어볶음"]

    for food in foods:
        f.write(food + "\n")

    print("food_list.txt 작성 완료")
```

```python
with open("./food_list.txt", "r", encoding="utf-8") as f:
    # print(f.read())

    foods = f.readlines()
    print(foods)

    if "김치찌개\n" in foods:
        idx = foods.index("김치찌개\n")
        foods[idx] = "순두부찌개\n"
    # print(idx)

print(foods)

with open("./food_list.txt", "w", encoding="utf-8") as f:
    f.writelines(foods)
```

## Comma Spreaded Value

```python
import csv

with open("./test_csv.csv", "w", encoding = "utf-8-sig") as f:
    writer = csv.writer(f)
    writer.writerow(("학번", "이름", "전화번호"))
    writer.writerows([(1, "홍길동", "010-1234-1234"), (2, "장보고", "010-0000-0000")])
```

```python
with open("./test_csv.csv", "r", encoding = "utf-8-sig") as f:
    reader = csv.reader(f)
    for i in reader:
        print(i)
```
