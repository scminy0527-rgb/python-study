```python
class User:
    name: str
    age: int
    address: str
    phone: str
    
    def __init__(self, name, age, address, phone):
        self.name = name
        self.age = age
        self.address = address
        self.phone = phone

    def __str__(self):
        return f"User(name: {self.name}, age: {self.age}, address: {self.address}, phone: {self.phone})"

    def print_hobby(self, hobby: str) -> None:
        print(f"{self.name}의 취미는 {hobby} 입니다.")
```

```python
jang = User("장보고", 20, "서울", "010-4567-4567")
```

```python
print(jang)
```

```python
from pydantic import BaseModel
```

```python
class Animal(BaseModel):
    name: str
    age: int
    feed: str
    hobby: str
    master: str
    gender: str
```

```python
cat = {
    "name": "고양이",
    "age": 20,
    "feed": "밥",
    "hobby": "잠자기",
    "master": "나",
    "gender": "남자"
}
```

```python
ddd = Animal(**cat)
```

```python
ddd
```

```python
class Employee(BaseModel):
    id: int
    name: str
    income: int
```

```python
employee1 = Employee(
    id = 1,
    name = "홍길동",
    income = 1000000
)
```

```python
employee2 = Employee(
    id = 2,
    name = "장보고",
    income = 1500000
)
```

```python
employee1
```

```python
employee2
```

```python
employee2.__eq__
```

```python
class Student(BaseModel):
    id: int
    name: str
    income: int
```

```python
student_data1 = {
    "id": 1,
    "name": "김영희",
    "income": 50000
}
```

```python
student_data2 = {
    "id": 2,
    "name": "김철수",
    "income": 100000
}
```

```python
student1 = Student(**student_data1)
```

```python
student2 = Student(**student_data2)
```

```python
student1
```

```python
student2
```

```python
# 기본적으로 주소값 비교

print(Employee.__eq__(employee1, employee2))
```

```python
student3 = Student(**student_data1)
```

```python
Employee.__eq__(student1, student3)
```
